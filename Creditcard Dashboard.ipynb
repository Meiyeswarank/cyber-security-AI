```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['V1'],df['V2'],color="red")
plt.title("V1 vs V2" )
plt.xlabel("V1")
plt.ylabel("V2")
plt.show()
```

```python
class_counts = df['Class'].value_counts()
plt.bar(class_counts.index, class_counts.values, color="blue")
plt.title("Class vs Count ")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
plt.plot(class_counts.index, class_counts.values, color="green")
plt.title("Class vs Count")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['Class'],color="orange",bins=4)
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python

plt.figure(figsize=(15,12))

plt.subplot(2, 2, 1)
plt.scatter(df['V1'], df['V2'], color="red")
plt.title("V1 vs V2")
plt.xlabel("V1")
plt.ylabel("V2")

plt.subplot(2, 2, 2)
class_counts = df['Class'].value_counts()
plt.bar(class_counts.index, class_counts.values, color="blue")
plt.title("Class vs Count")
plt.xlabel("Class")
plt.ylabel("Count")

plt.subplot(2, 2, 3)
plt.plot(class_counts.index, class_counts.values, color="green")
plt.title("Class vs Count")
plt.xlabel("Class")
plt.ylabel("Count")

plt.subplot(2, 2, 4)
plt.hist(df['Class'], color="orange", bins=4)
plt.title("Class Histogram")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```
